### This part will analysis the blogs people posted and give a score to evaluate it's a positive or nagetive blog 

we use this online api to do analysis https://bosonnlp.com/

```python
from __future__ import print_function, unicode_literals
import json
import requests
import xlrd
import openpyxl
```

```python
# change the fileName to analysis different files
# this path should direct to your root path, otherwise new file won't be saved!!!!  
fileName = r'/Users/young/Desktop/Data-si/1.FinalProject/Senti_analysis/Air_Quality/Beijing/Key-Word Beijing Air PM2.5.xlsx'

SENTIMENT_URL = 'http://api.bosonnlp.com/sentiment/analysis'
# Changing X-Token as user
headers = {
    'X-Token': '8sFM66ob.34061.X0YBbv8pj0AY',
    'Content-Type': 'application/json'
}

#  sheet name depends on the actual sheet
wb = xlrd.open_workbook(fileName)
sheet1 = wb.sheet_by_name('Sheet1')
rowNum = sheet1.nrows
x = rowNum / 100
#print(x)

#  according to the table, fix the index of columns
data_list = []
data_list.extend(sheet1.col_values(2))
# for item in data_list:
#     print(item)
data_res = []  #  save all the results from sentiment analysis in the While loop
n = 0
while n < x:
    #print(data_list[(n*10+1):((n+1)*10+1)])
    data = json.dumps(data_list[(n*100+1):((n+1)*100+1)])
    resp = requests.post(SENTIMENT_URL, headers=headers, data=data.encode('utf-8'))
    print(resp.text) 
    data_res.extend(eval(resp.text))
    n += 1

#‘_senti’  data written in a new excel , file name end with "_senti"
newFileName = fileName[0:len(fileName)-5] + "_senti" + fileName[len(fileName)-5:len(fileName)]


# date information saved in data_time[], if read Date format, transfer to String
data_time = []
for i in range(rowNum):
    if sheet1.cell(i, 0).ctype == 3:

        date = xlrd.xldate.xldate_as_datetime(sheet1.cell(i, 0).value, 0)
        dateStr = str(date.year)+u'/'+str(date.month)+u'/'+str(date.day)
        #print(dateStr)
        data_time.append(dateStr)

    else:
        data_time.append(sheet1.cell(i, 0).value)
    i += 1

data = []  # saving all the data in new excel

# saving process
counter = 0
while counter < len(data_res):
    data.append([data_time[counter + 1], data_res[counter][0], data_res[counter][1]])
    print(data[counter])
    counter += 1

title = ['time', 'positive', 'negative']
wb = openpyxl.Workbook()
ws1 = wb.active
ws1.append(title)
for row in data:
    ws1.append(row)
# the file will be saved in the same path with original file
wb.save(newFileName)
```
